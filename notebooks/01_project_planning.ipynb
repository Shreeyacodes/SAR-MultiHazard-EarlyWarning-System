# SAR-Based Multi-Hazard Early Warning System

## Phase 1: Flood Detection using Sentinel-1 SAR Data

```python
## 1. Problem Statement

Floods are one of the most destructive natural hazards, affecting human lives, agriculture, infrastructure, and ecosystems. This project aims to use Sentinel-1 SAR backscatter data and machine learning to detect flooded and non-flooded areas.
```

```python
## 2. Why SAR Data?

Synthetic Aperture Radar (SAR) can capture Earth surface information during day and night and can work through cloud cover. This makes SAR useful for flood detection, especially during extreme weather conditions.
```

```python
## 3. Machine Learning Task

This is a binary classification problem.

Classes:

- 0 = Non-flooded area
- 1 = Flooded area
```

```python
## 4. Input Features

Possible input features:

- VV backscatter
- VH backscatter
- VV/VH ratio
- Pre-flood backscatter
- Post-flood backscatter
- Backscatter difference
```

```python
## 5. Expected Output

The expected output is a flood classification result showing whether a given area/pixel/sample is flooded or non-flooded.
```
